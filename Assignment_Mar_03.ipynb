Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:
    
    Histograms:

    Uses: Histograms are useful for understanding the distribution of a single numerical variable. They provide insights into
        the central tendency, spread, and shape of the distribution of data.

    Scatter Plots:

    Uses: Scatter plots are used to visualize the relationship between two numerical variables. They can help identify patterns,
        correlations, and outliers in the data.

    Box Plots (or Whisker Plots):

    Uses: Box plots are excellent for visualizing the distribution of numerical data through their quartiles. They display the
        median, quartiles, and potential outliers, making them useful for detecting variations and understanding the spread and
        skewness of the data.

    Bar Plots (or Count Plots):

    Uses: Bar plots are ideal for comparing the frequency or count of categorical data across different categories. They allow
        you to visualize the distribution of categorical variables and compare the sizes of different groups or categories.

    Heatmaps:

    Uses: Heatmaps are graphical representations of data where individual values are represented as colors. They are 
        particularlyuseful for visualizing matrices or tables of data where colors indicate values, making it easier to
        identify patterns, correlations, or variations in data across two dimensions.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset using Seaborn
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn's lineplot function
plt.figure(figsize=(10, 6))  # Set the figure size
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri_data, ci=None)  # Line plot based on 'event'
# sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri_data, ci=None)  # Uncomment for line plot based on 'region'
plt.title('Line Plot of Signal vs. Timepoint for Different Events')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')  # Legend title based on 'event'; change if using 'region'
plt.grid(True)  # Add gridlines
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset using Seaborn
titanic_data = sns.load_dataset("titanic")

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns for two box plots

# First box plot for 'age' vs. 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Box Plot of Age vs. Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Second box plot for 'fare' vs. 'pclass'
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Box Plot of Fare vs. Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset using Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram for the 'price' column with hue based on 'cut'
plt.figure(figsize=(10, 6))  # Set figure size
sns.histplot(data=diamonds_data, x='price', hue='cut', element='step', kde=True, palette='viridis', multiple="stack")
plt.title('Histogram of Price with Different Cuts of Diamonds')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')  # Add legend with title 'Cut'
plt.grid(True)  # Add gridlines
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the "iris" dataset.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset using Seaborn
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue based on 'species'
sns.pairplot(iris_data, hue="species", palette="viridis", diag_kind="hist", markers=["o", "s", "D"])

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset using Seaborn
flights_data = sns.load_dataset("flights")

# Convert the dataset into a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create a heatmap using Seaborn's heatmap function
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)  # Annotate cells with passenger counts, format as integer
plt.title('Heatmap of Flight Passengers by Year and Month')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
